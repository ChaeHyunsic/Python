### 문제7
2000년대 영화들 중 Ratio가 2.0이상인 영화들을 Ratio가 높은 순으로 출력하시오. 


```
movie.where('Year', are.above_or_equal_to(2000)).where('Ratio', are.above_or_equal_to(2.0)).sort('Ratio', descending=True).show()
```

### 문제8
a와 b를 입력받아 Year가 a이상이고 Ratio가 b이상인 영화의 편수를 리턴하는 함수인 count_movies 함수를 작성하시오.
(Hint: Table()객체에는 num_rows와 num_columns라는 속성값을 통해 행과 열의 개수를 각각 표현한다.)

예) print(movie.num_rows)


```
def count_movies(a,b):
  result = movie.where('Year', are.above_or_equal_to(a)).where('Ratio', are.above_or_equal_to(b)).num_rows
  
  return result
```

```
count_movies(2000, 2.0)
```

### 보너스문제
위 데이터에 대하여 본인이 출제하고 싶은 문제가 있으면 한 문제를 출제하고 아래에 답안 코드를 작성하시오.

a, b, c를 입력받아 Year가 a이상 b이하에 있고 Ratio값이 c 이상인 영화의 Title을 리턴하는 함수인 whatis_return 함수를 작성하시오.

```
def whatis_return(a,b,c):
  result = movie.where('Year', are.between_or_equal_to(a,b)).where('Ratio', are.above_or_equal_to(c)).select('Title')
  
  return result
```

```
whatis_return(1999,2019,2.0)
```
