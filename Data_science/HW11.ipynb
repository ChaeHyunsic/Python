### 다음 임의로 생성된 데이터인 ``dt`` 테이블에 대하여 아래 질문들에 답하시오. 

```
np.random. seed(0)
def rand_table(r, n=1000, mx=0, my=0, s=1):
    x = mx + s * np.random.normal(0, 1, n)
    z = np.random.normal(0, 1, n)
    y = r*x*s + (np.sqrt(1-r**2))*z
    #plots.scatter(x, y)
    t= Table().with_columns('Y', y, 'X', x)
    return t
dt = rand_table(0.5, 1000, 2, 4, 2)   
dt
```

### 문제1
두 변수 ``X``와 ``Y``에 대하여 각각의 평균과 표준편차를 구해 ``mean_x``, ``std_x``,  ``mean_y``, ``std_y``변수에 저장하시오. 

```
mean_x = np.mean(dt.column(1))
std_x = np.std(dt.column(1))
mean_y = np.mean(dt.column(0)) 
std_y = np.std(dt.column(0))
```

```
mean_x, std_x, mean_y, std_y
```

### 문제2
두 변수 ``X``, ``Y``를 standard units으로 변환하여 ``ZX``, ``ZY`` 이름으로 dt테이블의 column으로 추가하시오.

```
dt = dt.with_columns(
    'ZY', (dt.column('Y') - mean_y)/std_y,
    'ZX', (dt.column('X') - mean_x)/std_x
)
```

```
dt
```

### 문제3
이번에는 두 변수 ``ZX``와 ``ZY``에 대하여 각각의 평균과 표준편차를 구해 ``mean_zx``, ``std_zx``,  ``mean_zy``, ``std_zy``변수에 저장하시오. 

```
mean_zx = np.mean(dt.column(3))
std_zx = np.std(dt.column(3))
mean_zy = np.mean(dt.column(2))
std_zy = np.std(dt.column(2))
```

```
mean_zx, std_zx, mean_zy, std_zy
```

### 문제4
두 변수 ``X``, ``Y`` 사이의 상관계수와 두 변수 ``ZX``, ``ZY`` 사이의 상관계수를 구하여 각각``corr_xy``와 ``corr_zxzy`` 변수에 저장하시오. 

```
corr_xy = np.mean(((dt.column('Y') - mean_y)/std_y)*((dt.column('X') - mean_x)/std_x));
corr_zxzy = np.mean(((dt.column('ZY') - mean_zy)/std_zy)*((dt.column('ZX') - mean_zx)/std_zx));
```

```
corr_xy, corr_zxzy
```

### 문제5
두 변수 ``X``, ``Y``사이와 ``ZX``, ``ZY`` 사이의 linear regression 결과 기울기 및 절편을 ``slope_xy``, ``intercept_xy``, ``slope_zxzy``, ``intercept_zxzy`` 변수에 차례대로 저장하시오.

(``Y``, ``ZY``가 y축변수 즉 종속변수이다)

```
slope_xy = corr_xy * np.std(dt.column('Y'))/np.std(dt.column('X'))
intercept_xy = np.mean(dt.column('Y')) - slope_xy*np.mean(dt.column('X'))
slope_zxzy = corr_zxzy * np.std(dt.column('ZY'))/np.std(dt.column('ZX'))
intercept_zxzy = np.mean(dt.column('ZY')) - slope_zxzy*np.mean(dt.column('ZX'))
```

```
slope_xy, intercept_xy, slope_zxzy, intercept_zxzy
```

### 보너스문제
위 데이터에 대하여 본인이 출제하고 싶은 문제가 있으면 한 문제를 출제하고 아래에 답안 코드를 작성하시오.

두 변수 X,Y사이와 ZX,ZY사이의 Linear Prediction를 구하여 Linear Prediction_xy, Linear Prediction_zxzy이름으로 dt테이블에 추가하시오.

또한, Linear Prediction에 대한 오차를 구하여 각각 Error_xy, Error_zxzy이름으로 dt테이블에 추가하시오.

```
dt = dt.with_column('Linear Prediction_xy',slope_xy*dt.column('X')+intercept_xy, 'Linear Prediction_zxzy',slope_zxzy*dt.column('ZX')+intercept_zxzy)
dt = dt.with_column('Error_xy',dt.column('Y')-dt.column('Linear Prediction_xy'), 'Error_zxzy',dt.column('ZY')-dt.column('Linear Prediction_zxzy'))
```
