# Data Indexing and Selection

NumPy array처럼 상당히 비슷하게 Pandas ``Series``와  ``DataFrame`` object 내 데이터에 접근한다.

## Data Selection in Series



### Series as dictionary



```python
import pandas as pd
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data
```

```python
data['b']
```

```python
'a' in data
```

```python
data.keys()
```

```python
list(data.items())
```

```python
data['e'] = 1.25
data
```

### Series as one-dimensional array

NumPy array의 *slices*, *masking*,  *fancy indexing* 등이 가능하다. 

```python
# slicing by explicit index
data['a':'c']
```

주의) 'c'  index 결과가 *포함* 된다. 

```python
# slicing by implicit integer index
data[0:2]
```

```python
# masking
data[(data > 0.3) & (data < 0.8)]
```

```python
# fancy indexing
data[['a', 'e']]
```

### Indexers: loc, iloc, and ix

정수 index가 사용되는 경우 아래 예처럼 explicit index(아래 data[1])인지  implicit Python-style index(아래 ``data[1:3]``)인지 애매한 경우가 생긴다. 

```python
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
data
```

```python
# explicit index when indexing
data[1]
```

```python
# implicit index when slicing
data[1:3]
```

이러한 혼돈을 피하기 위해 ``Series``에는 다음과 같은 attribute들을 제공한다.


먼저, ``loc`` explicit index로 접근을 한다.

```python
data.loc[1]
```

```python
data.loc[1:3]
```

반면, ``iloc`` attribute implicit Python-style index 형태로 접근한다.

```python
data.iloc[1]
```

```python
data.iloc[1:3]
```

``ix``은 혼합된 형태이며  ``DataFrame`` object에서 설명한다. 

## Data Selection in DataFrame


### DataFrame as a dictionary



```python
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})
data = pd.DataFrame({'area':area, 'pop':pop})
data
```

다음과 같이 각 column을 나타내는 ``Series`` object에 접근할 수 있다. 

```python
data['area']
```

attribute-style access 도 가능하다. 

```python
data.area
```

```python
data.area is data['area']
```

주의) column name이 문자열이 아니거나 이미 그런 attribute name이 존재하는 경우 사용할 수 없다. 

```python
data.pop is data['pop']
```

Dictionary-style access가 보다 안전하며 새로운 column으로 추가시킬 수도 있다. 

```python
data['density'] = data['pop'] / data['area']
data
```

### DataFrame as two-dimensional array

``values`` attribute를 통해 순수 데이터 array를 access 할 수 있다.

```python
data.values
```

전체 `DataFrame``의 row와 column들을 swap할 수 있다.

```python
data.T
```

NumPy array 에서의 작업이 또한 편리할 수도 있다. 

```python
data.values[0]
```

indexing의 혼돈을 피하기 위해 ``loc``, ``iloc``,  ``ix`` attribute를 사용한다. 

```python
data.iloc[:3, :2]
```

``loc``을 사용하여 동일한 결과를 만들 수 있다. 

```python
data.loc[:'Illinois', :'pop']
```

``ix``은 혼합된 방법이다. 

```python
data.ix[:3, :'pop']
```

이러한 것들과 기존 NumPy-style access와 결합될 수 있다. 
아래 예는 ``loc``와 masking 및 fancy indexing이 결합된 것 이다. 

```python
data.loc[data.density > 100, ['pop', 'density']]
```

Data 변경도 물론 이뤄진다. 

```python
data.iloc[0, 2] = 90
data
```

### Additional indexing conventions

반복적으로 다음 몇 가지 관례적 사용법을 눈여겨 보자.

```python
data['Florida':'Illinois']
```

```python
data[1:3]
```

```python
data[data.density > 100]
```
