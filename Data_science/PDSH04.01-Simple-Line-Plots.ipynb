# Simple Line Plots

$y = f(x)$ 함수를 그려보자. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

많은 경우 다음과 같이 figure 와 axes 객체를 생성하면서 시작한다.

```python
fig = plt.figure()
ax = plt.axes()
```

*figure* (an instance of the class ``plt.Figure``) object는 모든 그림 요소들(axes, graphics, text, labels)을 담고 있는 container이다.

*axes* (an instance of the class ``plt.Axes``) object는 위 그림과 같이 ticks 과 labels등을 표현한 사각형 박스를 표현한다.  ``ax.plot`` function 호출을 통해 그림을 그리게 된다. 

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

앞서 설명하였듯이 다음과 같은 형태로도 실행할 수 있다. (MATLAB-style)

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## Adjusting the Plot: Line Colors and Styles

``plt.plot()`` function 에서 여러 option이 있는데 그 중 몇가지를 살펴보자.

색상변경 - ``color`` keyword

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

line style -  ``linestyle`` keyword

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

위 2가지를 동시에 표현할 수 있다. (a single non-keyword argument)

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## Adjusting the Plot: Axes Limits

- axis limits : ``plt.xlim()`` and ``plt.ylim()`` methods

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

축을 거꾸로 배치할 수도 있다. 

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

양 축을 한꺼번에 처리하는 경우 - ``plt.axis()`` method (``[xmin, xmax, ymin, ymax]``을 전달)

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

또한 ``plt.axis()`` method는 축에 관한 layout을 설정하게 해 준다.

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## Labeling Plots

Titles 과 axis labels 

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

The position, size, and style of these labels can be adjusted using optional arguments to the function.
For more information, see the Matplotlib documentation and the docstrings of each of these functions.

``plt.legend()`` method

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

## Aside: Matplotlib Gotchas

위에서는 MATLAB-style interface을 설명하였다.

이를 object-oriented interface로 바꿀 경우 ``plt`` function와 동일한 이름의 ``ax`` method가 대부분 존재하므로 그렇게 변경하면 된다. 혹은 약간의 다른 이름이 사용되기도 한다. 

- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

object-oriented interface에서는 또한 이러한 함수들을 하나하나 호출하기 보다는 다음 예와 같이 ``ax.set()`` method에 한꺼번에 전달하면 편리하다. 

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
