# Vectorized String Operations

string(문자열)을 다루는 Pandas 기능을 살펴보자. 특히 vectorized operation으로 구현되어 효율적이다. 

## Introducing Pandas String Operations


```python
import numpy as np
x = np.array([2, 3, 5, 7, 11, 13])
x * 2
```

위와 같이 NumPy 숫자 array의 경우 효과적인 vector operation이 제공된다.
그러나 문자열인 경우에는 그렇지 못하므로 loop를 통해 접근해야 한다. 

```python
data = ['peter', 'Paul', 'MARY', 'gUIDO']
[s.capitalize() for s in data]
```

더구나 missing value에 대한 처리도 주의해야 한다. 

```python
data = ['peter', 'Paul', None, 'MARY', 'gUIDO']
[s.capitalize() for s in data]
```

Pandas에서는 ``str``속성을 사용하여 vector operation과 missing value에 대한 처리를 기본적으로 제공한다. 

```python
import pandas as pd
names = pd.Series(data)
names
```

```python
names.str.capitalize()
```

``str`` 속성을 이용한 많은 다른 함수들이 포함되어 있다.

## Tables of Pandas String Methods



```python
monte = pd.Series(['Graham Chapman', 'John Cleese', 'Terry Gilliam',
                   'Eric Idle', 'Terry Jones', 'Michael Palin'])
```

### Methods similar to Python string methods
거의 모든  Python built-in 문자열  method에 대응하는 Pandas Series 객체내``str`` method가 있다. 

|             |                  |                  |                  |
|-------------|------------------|------------------|------------------|
|``len()``    | ``lower()``      | ``translate()``  | ``islower()``    | 
|``ljust()``  | ``upper()``      | ``startswith()`` | ``isupper()``    | 
|``rjust()``  | ``find()``       | ``endswith()``   | ``isnumeric()``  | 
|``center()`` | ``rfind()``      | ``isalnum()``    | ``isdecimal()``  | 
|``zfill()``  | ``index()``      | ``isalpha()``    | ``split()``      | 
|``strip()``  | ``rindex()``     | ``isdigit()``    | ``rsplit()``     | 
|``rstrip()`` | ``capitalize()`` | ``isspace()``    | ``partition()``  | 
|``lstrip()`` |  ``swapcase()``  |  ``istitle()``   | ``rpartition()`` |


```python
monte.str.lower()
```

```python
monte.str.len()
```

```python
monte.str.startswith('T')
```

```python
monte.str.split()
```

### Methods using regular expressions

또한 정규표현식(RE; regular expression)을 지원하며 하며 Python's built-in ``re`` module에서 제공하는 방식을 많이 따른다. 

| Method | Description |
|--------|-------------|
| ``match()`` | Call ``re.match()`` on each element, returning a boolean. |
| ``extract()`` | Call ``re.match()`` on each element, returning matched groups as strings.|
| ``findall()`` | Call ``re.findall()`` on each element |
| ``replace()`` | Replace occurrences of pattern with some other string|
| ``contains()`` | Call ``re.search()`` on each element, returning a boolean |
| ``count()`` | Count occurrences of pattern|
| ``split()``   | Equivalent to ``str.split()``, but accepts regexps |
| ``rsplit()`` | Equivalent to ``str.rsplit()``, but accepts regexps |

따라서 문자열 처리가 요구되는 경우 유용하다. 

```python
monte.str.extract('([A-Za-z]+)', expand=False)
```

문자열의 처음과 끝을 표현하는 (``^``) 과 (``$``) 등을 사용해 보자. 

```python
monte.str.findall(r'^[^AEIOU].*[^aeiou]$')
```
