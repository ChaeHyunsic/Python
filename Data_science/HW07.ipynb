### 문제1
2명의 사람이 가위바위보를 100번 시행하여 비긴 횟수를 리턴하는 함수 count_tie_100()를 작성하시오. 

(단, 위에서 정의한 rcp 배열을  이용하여 작성할 것)

```
def count_tie_100():
  count=0
  
  for i in range(100):
    alpha =  np.random.choice(rcp)
    beta =  np.random.choice(rcp)
    if(alpha == beta):
      count+=1
      
  return count
```

### 문제2
2명의 사람이 가위바위보를 ``no_rcp``번 시행하여 비긴 횟수를 리턴하는 함수 count_tie()를 작성하시오. 

(단, ``no_rcp``은 입력 파라미터이며 default는 100으로 설정한다.)

```
def count_tie(no_rcp=100):
  count=0
  
  for i in range(no_rcp):
    alpha =  np.random.choice(rcp)
    beta =  np.random.choice(rcp)
    if(alpha == beta):
      count+=1
      
  return count
```

### 문제3
2명이 ``no_rcp``번 랜덤하게 가위바위보 했을 때 비긴 ***비율***(주의: 횟수가 아님)을 조사하는 시행을 ``no_trial``번 반복하여 그 결과(비긴 비율)를 배열로 리턴하는 함수 ratio_tie를 작성하시오.

(``no_rcp``과 ``no_trial``은 입력 파라미터이며 default값은 100, 10000으로 각각 설정한다.)

```
def ratio_tie(no_rcp=100, no_trial=10000):
  ratio=np.arange(no_trial,dtype='float_')
  
  for i in range(no_trial):
    count=0
    for j in range(no_rcp):
      alpha =  np.random.choice(rcp)
      beta =  np.random.choice(rcp)
      if(alpha == beta):
        count+=1
    ratio[i]=float(count)/float(no_rcp)
  
  return ratio.round(2)
```

```
ratio_tie(50, 5000)
```

```
ratio_tie()
```

### 보너스문제
위 데이터에 대하여 본인이 출제하고 싶은 문제가 있으면 한 문제를 출제하고 아래에 답안 코드를 작성하시오.

3명이 no_rcp번 랜덤하게 가위바위보 했을 때 모두 비길 비율(주의: 횟수가 아님)을 조사하는 시행을 no_trial번 반복하여 그 결과(비율)를 배열로 리턴하는 함수 three_ratio_tie를 작성하시오. (no_rcp, no_trial은 입력 파라미터이며 default값은 100, 10000으로 각각 설정한다.)

```
def three_ratio_tie(no_rcp=100, no_trial=10000):
  ratio=np.arange(no_trial,dtype='float_')
  
  for i in range(no_trial):
    count=0
    for j in range(no_rcp):
      alpha = np.random.choice(rcp)
      beta = np.random.choice(rcp)
      gamma = np.random.choice(rcp)
      if((alpha == beta) and (alpha == gamma) and (beta == gamma)):
        count+=1
    ratio[i]=float(count)/float(no_rcp)
  
  return ratio.round(2)
```

```
three_ratio_tie(200,5000)
```
