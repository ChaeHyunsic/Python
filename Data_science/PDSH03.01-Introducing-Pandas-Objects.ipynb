# Introducing Pandas Objects

먼저 3가지 Pandas 기본 구조인 ``Series``, ``DataFrame``, ``Index``을 살펴보기 위해 다음 두 패키지들을 import하자. 

````python
import numpy as np
import pandas as pd
````

## The Pandas Series Object

``Series``는 1차원 array이며 list나 array를 통해 생성할 수 있다. 

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
````

아래 두 결과에서 보듯이 ``Series`` object는 ``values`` 와 ``index``라는 속성(attributes)이 있다.
이 때, ``values``는 앞서 배운 NumPy array이다. 

````python
data.values
````

````python
data.index
````

NumPy array처럼 기본적으로 simple index와 slicing으로 데이터를 접근할 수 있다.

````python
data[1]
````

````python
data[1:3]
````

### ``Series`` as generalized NumPy array

``Series`` object 는 기본적으로 one-dimensional NumPy array이다.

그러나, 차이점은 index의 유무이다.
- Numpy array : *implicitly defined* integer index 
- Pandas ``Series`` : *explicitly defined* index

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data
````

따라서 정수(integer)가 아닌 문자를 통해 indexing이 가능하다. 

````python
data['b']
````

또한 정수인 경우도 연속값일 필요가 없다. 

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=[2, 5, 3, 7])
data
````

````python
data[5]
````

### Series as specialized dictionary

분명 Pandas ``Series``는 Python dictionary와 유사해 보인다. 하지만
- dictionary는 ``arbitrary key``로부터 ``arbitrary value``로의 map
- ``Series``는 ``typed keys``로부터 ``typed value`` 로의 map

다시 말해 `Series`가 더 빠르게 실행될 수 있다. 

````python
population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)
population
````

default로 ``Series``는 sorted key 형태로 index를 생성한다.

````python
population['California']
````

dictionary와 달리 ``Series``는 array-style indexing을 지원한다. (예: slicing)

````python
population['California':'Illinois']
````

### Constructing Series objects

Recap) Pandas ``Series`` 생성 문법

```python
>>> pd.Series(data, index=index)
```

이 때, ``data``는 여러 형태의 데이터이지만 주로 list나 NumPy array와 같은 종류이며,  ``index``는 optional argument로서 default는 integer sequence이다.

````python
pd.Series([2, 4, 6])
````

``data`` 는 scalar 값으로서 반복 채움형태로 사용될 수 있다. 

````python
pd.Series(2, index=[100, 200, 300])
````

``data`` dictionary일 수 있으며 그러면  ``index``는  sorted dictionary key로 정의된다.

````python
pd.Series({2:'a', 1:'b', 3:'c'})
````

``index``가 명시적으로 한정될 수 있다.

````python
pd.Series({2:'a', 1:'b', 3:'c'}, index=[3, 2])
````

## The Pandas DataFrame Object



### DataFrame as a generalized NumPy array
``DataFrame``은 NumPy의 two-dimensional array에 비유될 수 있으며 보다 flexible row indices 와 flexible column names을 갖는다.

``DataFrame``은 index를 공유하는 ``Series`` object들의 모임이라 할 수 있다.

````python
area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict)
area
````

앞서 ``population`` Series 와 index를 공유하는 DataFrame을 만들 수 있다. 

````python
states = pd.DataFrame({'population': population,
                       'area': area})
states
````

``Series``처럼 ``DataFrame``은 ``index`` attribute가 있다. 

````python
states.index
````

추가적으로 ``DataFrame``은 ``columns`` attribute가 존재한다.

````python
states.columns
````

````python
states['area']
````

### Constructing DataFrame objects

``DataFrame``은 아래 예들처럼 다양한 방식으로 생성할 수 있다. 

#### From a single Series object



````python
pd.DataFrame(population, columns=['population'])
````

#### From a list of dicts


````python
data = [{'a': i, 'b': 2 * i}
        for i in range(3)]
pd.DataFrame(data)
````

dictionary내 일부 key의 값이 없어도 Pandas는 그것을 ``NaN`` (i.e., "not a number") 으로 채운 후 DataFrame을 생성한다.

````python
pd.DataFrame([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}])
````

#### From a dictionary of Series objects


````python
pd.DataFrame({'population': population,
              'area': area})
````

#### From a two-dimensional NumPy array


````python
pd.DataFrame(np.random.rand(3, 2),
             columns=['foo', 'bar'],
             index=['a', 'b', 'c'])
````

#### From a NumPy structured array



````python
A = np.zeros(3, dtype=[('A', 'i8'), ('B', 'f8')])
A
````

````python
pd.DataFrame(A)
````

## The Pandas Index Object

``Index`` object에 관련된 기능들을 살펴보자. 

````python
ind = pd.Index([2, 3, 5, 7, 11])
ind
````

### Index as immutable array



````python
ind[1]
````

````python
ind[::2]
````

````python
print(ind.size, ind.shape, ind.ndim, ind.dtype)
````

``Index`` object는 수정될 수 없는 immutable array이다. 

````python
ind[1] = 0
````

### Index as ordered set

``Index`` object는 Python's built-in ``set`` data structure를 따르므로 unions, intersections, differences 등 집합 연산이 가능하다. 

````python
indA = pd.Index([1, 3, 5, 7, 9])
indB = pd.Index([2, 3, 5, 7, 11])
indA+indB
````

````python
indA & indB  # intersection
````

````python
indA | indB  # union
````

````python
indA ^ indB  # symmetric difference
````

사실 이러한 집합 연산은 object method로도 가능하다.
예) ``indA.intersection(indB)``.
