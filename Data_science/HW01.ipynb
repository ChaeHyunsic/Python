```
import numpy as np
rng=np.random.RandomState(100)
```

### 문제1
0이상 100이하 정수들로 구성되는 100행 5열의 2차원 array인 ``score``를 생성하시오.
(단, 위 rng를 사용할 것)

```
score=rng.randint(100,size=(100,5))
score
```

### 문제2
지금부터 ``score``가 100명의 학생들의 국어, 영어, 수학, 과학, 사회 점수를 뜻한다고 가정하자.

이제 5개 과목의 평균점수를 뜻하는 ``avgx`` array(1차원)를 생성하시오. 

```
avgx_array=score.mean(axis=0)
avgx_array
```

### 문제3
학생별 평균점수를 뜻하는 ``avgy`` array(1차원)를 생성하시오. 

```
avgy_array=score.mean(axis=1)
avgy_array
```

### 문제4
100명 중 홀수번째는 여학생, 짝수번째는 남학생 정보를 뜻한다.

여학생 전체의 5개과목별 평균점수를 뜻하는 ``avgx1`` array(1차원)을 구하시오.

```
avgx1_array=score[0::2].mean(axis=0)
avgx1_array
```

### 문제5
마찬가지로 남학생 전체의 5개과목별 평균점수를 뜻하는 ``avgx2`` array(1차원)을 구하시오.

```
avgx2_array=score[1::2].mean(axis=0)
avgx2_array
```

### 문제6

$\frac{avgx1+avgx2}{2}$ 결과를 출력하시오. (변수에 저장할 필요 없음)

(그리고 출력된 array가 ``avgx1``과 같은지 확인)

```
(avgx1_array+avgx2_array)/2
```

### 문제7
각 점수가 해당과목 평균으로부터 얼마나 차이가 나는지를 나타내는 ``score_dev`` array(2차원)를 생성하시오.
- 즉, score_dev의 shape는 score의 shape와 동일하다.
- 점수가 30점이고 평균이 45점이면 그 차이는 $30-45=-15$ 점이다.


```
score_dev_array=score-avgx_array
score_dev_array
```

### 문제8
``score``에서 ``avgx``를 계산하였듯이 ``score_dev``에서 axis=0 방향으로 평균을 나타내는 ``scorex_dev``를 구하시오. 

```
scorex_dev=score_dev_array.mean(axis=0)
scorex_dev
```

### 문제9
전체과목 평균을 기준으로 1등을 한 학생의 5개 과목 점수를 출력하시오. (1차원 array 형태로 출력하면 됨) 

```
winner=score[avgy_array==avgy_array.max()]
winner
```

### 문제10
모든 과목이 각 과목의 평균점수보다 높은 학생수를 구하시오. 

```
member=np.sum(np.sum(score_dev_array>0,axis=1)==5,axis=0)
member
```

### 보너스문제
위 데이터에 대하여 본인이 출제하고 싶은 문제가 있으면 한 문제를 출제하고 아래에 답안 코드를 작성하시오.

모든 과목을 고르게 공부하여 가장 성적이 고르게 나온 학생을 구하시오.(표준편차를 이용)
