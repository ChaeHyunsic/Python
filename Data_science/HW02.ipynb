```
import numpy as np
import pandas as pd
```

다음은 100명 학생에 관한 정보를 담고 있는 변수들이다. 
 - sname : 학생 이름
 - noclass : 반 번호
 - gender : 성별
 - height : 키
 - weight : 몸무게
 
아래 코드는 수정없이 사용하면 된다.

```
sname=['s'+format(i, '03d') for i in range(100)]
noclass=np.random.randint(1, 6, 100)
gender=['female']*50 + ['male']*50
height=np.random.randint(160, 190, 100)
weight=np.random.randint(50, 80, 100)
```

### 문제1
sname, noclass, gender, height, weight를 column으로 사용하는 DataFrame 변수 student를 생성하시오. 
(index는 자동으로 생성되는 정수)

```
student=pd.DataFrame({'gender':gender,'height':height,'noclass':noclass,'sname':sname,'weight':weight})
```

```
student.head()  ### 첫 5행을 출력한 결과이다.
```

### 문제2
1반부터 5반까지 담임교사의 성이 각각 Kim, Lee, Park, Choi, Yun이라고 할 때 반번호(1반~5반)와 담임교사의 이름(tname)을 column으로 가지는 DataFrame 변수 teacher 를 생성하시오.  (index는 자동으로 생성되는 정수값이다)

```
tname=pd.Series(['kim','Lee','Park','Choi','Yun'])
noclass=pd.Series([1,2,3,4,5])
teacher=pd.DataFrame({'tname':tname,'noclass':noclass})
```

```
teacher
```

### 문제3
student와 teacher를 합쳐서 info 라는 이름의 DataFrame변수를 생성하시오. 

```
info=pd.merge(student,teacher)
```

```
info.tail() ### 마지막 5행을 보여준다.
```

### 문제4
$ratio=height-weight$ 일 때 새로운 column ``ratio``를 info 에 추가하시오. 

```
ratio=pd.DataFrame({'ratio':height-weight})
info=info.join(ratio)
```

```
info.head()
```

### 문제5
info에서 sname변수가 index로 되도록 변경한 info_sname 변수를 생성하시오.


```
info_sname=info.set_index('sname')
```

```
info_sname.head()
```

### 문제6
반별로 height, weight, ratio의 평균값을 출력하시오. 
(info와 info_sname 중 어떤 변수를 사용하여도 좋다)

```
info.groupby('noclass').mean()
```

### 문제7
pivot table을 사용하여 반별, 성별  weight 평균값을 출력하시오.

```
info.pivot_table('weight', index='noclass', columns='gender')
```

### 문제8
각 반에서 남학생과 여학생 수를 가리키는 DataFrame 변수 ct를 생성하시오.

(단, 이 때 noclass와 gender 순서로 multi-index하며 column이름은 count로 한다)

```
ct=info.groupby(['noclass','gender']).count().drop('height',axis=1).drop('sname',axis=1).drop('weight',axis=1).drop('tname',axis=1).rename(columns={'ratio':'count'})
```

```
ct
```

### 보너스문제
위 데이터에 대하여 본인이 출제하고 싶은 문제가 있으면 한 문제를 출제하고 아래에 답안 코드를 작성하시오.

담임 교사에 따른 ratio분포를 살펴보려고 한다. 이를 부합하는 DataFrame 변수 rt를 생성하시오.
(단, 이 때 tname과 ratio 순서로 multi-index하며 column이름은 state로 한다)

```
rt=info.groupby(['tname','ratio']).count().drop('gender',axis=1).drop('height',axis=1).drop('noclass',axis=1).drop('sname',axis=1).rename(columns={'weight':'state'})
```
