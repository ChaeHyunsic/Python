# Simple Scatter Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Scatter Plots with ``plt.plot``

scatter(산점도) plot을 그리는 첫 번째 방법은 앞절에서 배운 ``plt.plot``/``ax.plot``에서 marker 문자를 추가하여  line 대신 point로 표현하는 것이다. 

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

다음은 몇가지 marker의 모양을 보인다.

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

line style/color 와 같이 사용되면 linepoint 그림이 된다. 

```python
plt.plot(x, y, '-ok');
```

다음 예와 같이 line 과 marker에 관한 여러 속성을 정의할 수 있다.

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

## Scatter Plots with ``plt.scatter``

scatter(산점도)를 그리는 다른 방법은 ``plt.scatter`` function를 이용하는 것이다. 

```python
plt.scatter(x, y, marker='o');
```

``plt.scatter``의 차별점은 각 point의 속성(size, face color, edge color, etc.)이 개별적으로 다르게 표현될 수 있다는 점이다. 


```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

위 예제에서는 target 정보를 이용함으로써 보다 유용한 시각화를 하게 된다.

## ``plot`` Versus ``scatter``: A Note on Efficiency

``plt.plot``이 ``plt.scatter``보다 빠르게 그린다.
따라서 대량의 데이터에 대해 산점도를 단순하게 그리는 경우라면 ``plt.plot``을 사용하는 것이 효율적이다.
