```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
```

다음은 100명 학생에 관한 정보를 담고 있는 변수들이다. 
 - years : 학년 (1~4)
 - gender : 성별 ('female', 'male')
 - height : 키
 - weight : 몸무게
 
아래 코드는 수정없이 사용하면 된다.

```
rng=np.random.RandomState(0)
years=rng.randint(1, 5, size=100)
gender=['female']*50 + ['male']*50
height=np.concatenate([rng.randn(50)*4+160, rng.randn(50)*7+170]) 
weight=np.concatenate([rng.randn(50)*4+60, rng.randn(50)*7+70]) 
```

### 문제1
heigth(x축)와 weight(y축) 사이의 산점도를 그리시오. 

(단, marker는 'o' 사용)

```
plt.plot(height,weight,'o')
```

### 문제2
처음 50명이 'female', 나머지 50명이 'male' 데이터이므로 이를 서로 다른 종류로 구분하여 산점도를 그리시오. 

단,  
- legend를 오른쪽아래 테두리 있는 형태로 추가
- marker는 'o'(female), 's'(male) 사용
- x축과 y축 label을 'height', 'weight'로 표시

```
plt.plot(height[:50],weight[:50],'o',label='female')
plt.plot(height[50:],weight[50:],'s',label='male')
plt.legend(loc='lower right')
plt.xlabel("height")
plt.ylabel("weight")
```

### 문제3
이번에는 ``years`` 변수 내용을 이용하여 학년별 산점도를 그리시오.

단,  
- legend를 왼쪽위 테두리 있는 형태로 추가하고 Year1, Year2, Year3, Year4로 label 부여
- marker는 'o'(Year1), 's'(Year2), 'd'(Year3), '+'(Year4) 사용
- x축과 y축 label을 'height', 'weight'로 표시

```
plt.plot(height[years==1],weight[years==1],'o',label='Years1')
plt.plot(height[years==2],weight[years==2],'s',label='Years2')
plt.plot(height[years==3],weight[years==3],'d',label='Years3')
plt.plot(height[years==4],weight[years==4],'+',label='Years4')
plt.legend(loc='upper left')
plt.xlabel("height")
plt.ylabel("weight")
```

### 문제4
학년별 height의 평균(mean)과 표준편차(std)를 이용하여 errorbar를 그리시오. 

단,
- errorbar 속성은 수업시간에 배운 fmt='-k' 사용
- x축과 y축 label을 'year', 'height'로 표시
- x축 범위는 0부터 5까지로 설정

```
eb = pd.Series(height,index=years)
x=[1,2,3,4]
plt.xlim(0,5)
dy = eb.groupby(years).std()
y = eb.groupby(years).mean()
plt.errorbar(x,y, yerr=dy, fmt='-k')
plt.xlabel("year")
plt.ylabel("height")
```

### 문제5
학년별 height와 weight의 histogram을 다음 조건에 맞게 그리시오. 
- 4행 2열의 subplot을 생성하여 각 행은 year를 뜻하고, 1번 및 2번 열은 각각 height와 weight 결과를 뜻한다.
- 모든 histogram에서 bin의 개수는 10이다.
- 전체 그림의 크기는 [6.4, 6.4]이다.
- subplot 사이의 공백은 hspace=0.4, wspace=0.4이다.


```
fig = plt.figure(figsize=(6.4, 6.4))
grid = plt.GridSpec(4, 2, wspace=0.4, hspace=0.4)

a_hist = fig.add_subplot(grid[:1,:1])
b_hist = fig.add_subplot(grid[:1,1:2])
c_hist = fig.add_subplot(grid[1:2,:1])
d_hist = fig.add_subplot(grid[1:2,1:2])
e_hist = fig.add_subplot(grid[2:3,:1])
f_hist = fig.add_subplot(grid[2:3,1:2])
g_hist = fig.add_subplot(grid[3:4,:1])
h_hist = fig.add_subplot(grid[3:4,1:2])

a_hist.hist(height[years==1],10, histtype='stepfilled',orientation='vertical')
b_hist.hist(weight[years==1],10, histtype='stepfilled',orientation='vertical')
c_hist.hist(height[years==2],10, histtype='stepfilled',orientation='vertical')
d_hist.hist(weight[years==2],10, histtype='stepfilled',orientation='vertical')
e_hist.hist(height[years==3],10, histtype='stepfilled',orientation='vertical')
f_hist.hist(weight[years==3],10, histtype='stepfilled',orientation='vertical')
g_hist.hist(height[years==4],10, histtype='stepfilled',orientation='vertical')
h_hist.hist(weight[years==4],10, histtype='stepfilled',orientation='vertical')
```

### 보너스문제
위 데이터에 대하여 본인이 출제하고 싶은 문제가 있으면 한 문제를 출제하고 아래에 답안 코드를 작성하시오.

성별 height와 weight의 histogram을 다음 조건에 맞게 그리시오.

*   2행 2열의 subplot을 생성하여 각 행은 gender를 뜻하고, 1번 및 2번 열은 각각 height와 weight 결과를 뜻한다.
     (처음 50명이 'female', 나머지 50명이 'male' 데이터)
*   모든 histogram에서 bin의 개수는 10이다.
*   전체 그림의 크기는 [6.4, 6.4]이다.
*   subplot 사이의 공백은 hspace=0.4, wspace=0.4이다.

```
fig = plt.figure(figsize=(6.4, 6.4))
grid = plt.GridSpec(4, 2, wspace=0.4, hspace=0.4)

alpha_a_hist = fig.add_subplot(grid[:1,:1])
alpha_b_hist = fig.add_subplot(grid[:1,1:2])
beta_c_hist = fig.add_subplot(grid[1:2,:1])
beta_d_hist = fig.add_subplot(grid[1:2,1:2])

alpha_a_hist.hist(height[:50],10, histtype='stepfilled',orientation='vertical')
alpha_b_hist.hist(weight[:50],10, histtype='stepfilled',orientation='vertical')
beta_c_hist.hist(height[50:],10, histtype='stepfilled',orientation='vertical')
beta_d_hist.hist(weight[50:],10, histtype='stepfilled',orientation='vertical')
```
